```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Maryland PUMA

 * NP: Number of household people 
 * HHT: Household or family type
 * HINCP: Household income 
 * HUPAC: Household presence and age of children 
 * WIF: Workers in family during the last 12 months
 * AGEP: Age of the person
 * SEX: Gender of the person
 * ESR: Employment status of the person 
 * RAC1P: Recorded detailed race 

## Puma 100

```python
data_path = "data/Sample 2010-1-PUMA100_Treated.csv"
df = pd.read_csv(data_path)
df.head(10)
```

```python
df.columns[1]
```

```python
df.describe()
```

## Distribution plots

```python
fig, ax = plt.subplots(3,3, figsize=(25, 17))
k = 0
for i in range(3):
    for j in range(3):
        col = df.columns[k]
        sns.distplot(df[col], ax=ax[i, j], label=col)
        k += 1
plt.show()
```

## Joint plots

Age VS Household income

```python
sns.jointplot(x="AGEP", y="HINCP", data=df)
plt.show()
```

Number of household people VS Household income

```python
sns.jointplot(x="NP", y="HINCP", data=df)
plt.show()
```

Age VS Employment status of the person

```python
sns.jointplot(x="AGEP", y="ESR", data=df)
plt.show()
```
